```python
import re, nltk
# nltk.download('stopwords' , r"C:/Users/chira/OneDrive/Desktop/ml-ai/twitter_sentiments/nltk_data")
from nltk.corpus import stopwords
# nltk.download('wordnet' , r"C:/Users/chira/OneDrive/Desktop/ml-ai/twitter_sentiments/nltk_data")
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
# nltk.download("wordnet", r"C:/Users/chira/OneDrive/Desktop/ml-ai/twitter_sentiments/nltk_data")
# nltk.data.path.append('./nltk_data/')

stop_words = set(stopwords.words('english'))
wordnet_lemmatizer = WordNetLemmatizer()

def normalizer(tweet):
    only_letters = re.sub("[^a-zA-Z]", " ",str(tweet))
    tokens = nltk.tokenize.word_tokenize(only_letters)[2:]
    lower_case = [l.lower() for l in tokens]
    filtered_result = list(filter(lambda l: l not in stop_words, lower_case))
    lemmas = [wordnet_lemmatizer.lemmatize(t) for t in filtered_result]
    return lemmas
```

```python
from nltk import ngrams
def ngrams(input_list):
    #onegrams = input_list
    bigrams = [' '.join(t) for t in list(zip(input_list, input_list[1:]))]
    trigrams = [' '.join(t) for t in list(zip(input_list, input_list[1:], input_list[2:]))]
    return bigrams+trigrams
tweets['grams'] = tweets.normalized_tweet.apply(ngrams)
tweets[['grams']].head()
```

```python
import collections
def count_words(input):
    cnt = collections.Counter()
    for row in input:
        for word in row:
            cnt[word] += 1
    return cnt
```

```python
import numpy as np
from scipy.sparse import hstack
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(ngram_range=(1,2))
```

```python
vectorized_data = count_vectorizer.fit_transform(tweets.clean_text.values.astype('str'))
indexed_data = hstack((np.array(range(0,vectorized_data.shape[0]))[:,None], vectorized_data))
```

```python
def label_target(tweets):
    if tweets.category == 'Negative':
        return 0
    elif tweets.category == "Neutral":
        return 1
    else:
        return 2

targets = tweets.apply(label_target, axis=1)
```

```python
targets
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test, targets_train, targets_test = train_test_split(indexed_data, targets, test_size=0.9, random_state=0)
data_train_index = data_train[:,0]
data_train = data_train[:,1:]
data_test_index = data_test[:,0]
data_test = data_test[:,1:]
```
